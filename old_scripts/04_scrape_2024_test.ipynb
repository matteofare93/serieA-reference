```python
competitions = json.loads(soup.find('script', {'type':'application/json'}).text)
competitions = competitions['props']['pageProps']['allCompetitionsBySeriesId']['competitions']
```

```python
year = '2023/2024'.split('/')[0]
this_competitions = [x for x in competitions if x['year'] == int(year)]
```

```python
params = {
    'selectedTeamId': 'all',
    'selectedTab': 'schedule',
    #'championshipTypeId': 0, # 2 is playoff, 4 is regular season
    'phaseToShow': 'all',
    #'matchDay': 0, # 1-30 for regular season, what about playoffs?
    'year': '2023' # year of the first part of the season
}
```

```python
for comp in this_competitions:
    
    comp_id = comp['championship_type_id']
    comp_name = comp['full_name']
    
    params['championshipTypeId'] = comp_id
    params['matchDay'] = 17
    
    query_string = urlencode(params)
    full_url = base_url + query_string

    user_agent = random.choice(user_agent_list)
    headers = {'User-Agent': user_agent}

    req = requests.get(full_url, headers=headers)
    soup = BS(req.text, 'html.parser')
    break
```
