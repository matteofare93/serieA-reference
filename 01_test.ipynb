```python
import requests
import time
from bs4 import BeautifulSoup as BS
import pandas as pd
```

```python
base_url = 'http://web.legabasket.it/team/counter/prandoni_bergamo/schedule'
base_url = base_url.replace('counter','1200')

print(base_url)
req = requests.get(base_url)
soup = BS(req.text, 'lxml')

col = ['Giornata','Data','Squadra Casa','Punteggio','Squadra Trasferta','Link Boxscore']
games_df = pd.DataFrame(columns=col)

table = soup.findAll('table')[0]
```

```python
soup
```

```python
soup.find('div', {'class':'title'}).get_text().strip().split(' ')[-1]
```

```python
pos = 0
for i,tr in enumerate(table.findAll('tr')):
    row = []
    for i,td in enumerate(tr.findAll('td')):
        try:
            a = td.findAll('a')
            texts = [x.text for x in a]
        except IndexError:
            continue
        row.extend(texts)
        if i == 2:
            link = a[0]['href']
    if len(row) > 0:
        row.append(link)
        games_df.loc[pos] = row
        pos += 1
```

```python
def convert_date(datetime):
    if len(datetime) == 11:
        return datetime[0:-1] + ' 00:00'
    else:
        return datetime[0:10] + ' ' + datetime[10:]
```

```python
games_df['Data'] = games_df['Data'].apply(convert_date)
```
