```python
import pandas as pd
import datetime
```

```python
df_raw = pd.read_csv('./out_data/all_games_with_team_id.csv')
df_raw.tail()
```

```python
df_raw.groupby('Link Boxscore').aggregate({'Team ID':'count'}).reset_index().sort_values('Team ID')
```

```python
df_clean = df_raw.drop_duplicates(subset=['Link Boxscore'], keep='first')
print(df_raw.shape[0], df_clean.shape[0])
```

```python
#df_clean = pd.to_datetime(df_clean['Data'], format='%d/%m/%Y %H:%M', errors='ignore')
#df_clean.head()
```

```python
df_clean.rename(columns={'Data':'DataOra'}, inplace=True)
```

```python
def split_data(dateora):
    return dateora.split(' ')[0]

def invert_data(data):
    day, month, year = data.split('/')
    return year + '/' + month + '/' + day

df_clean['Data'] = df_clean['DataOra'].apply(split_data)
df_clean['Data'] = df_clean['Data'].apply(invert_data)
df_clean.head()
```

```python
#df_clean['Giornata'] = df_clean['Giornata'].str.replace('°','')
#df_clean['Giornata'] = df_clean['Giornata'].str.replace('Giornata','')
#df_clean.head()
```

```python
df_clean.sort_values(['Anno','Data'])
```

```python
df_clean['Squadra Casa'] = df_clean['Squadra Casa'].str.strip()
df_clean['Squadra Trasferta'] = df_clean['Squadra Trasferta'].str.strip()
df_clean['Anno'] = df_clean['Anno'].str.strip()
```

```python
df_group = df_clean.groupby('Anno').aggregate({'Team ID':'count','Squadra Casa':pd.Series.nunique,'Squadra Trasferta':pd.Series.nunique}).reset_index()
```
